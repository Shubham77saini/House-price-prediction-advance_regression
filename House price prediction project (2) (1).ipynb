## Handling missing value of Bsmt feature

```python
bsmt_col = ['BsmtQual', 'BsmtCond', 'BsmtExposure','TotalBsmtSF','BsmtFinType1', 'BsmtFinSF1',
            'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'BsmtFullBath', 'BsmtHalfBath']
bsmt_feat = df[bsmt_col]
bsmt_feat
```

```python
bsmt_feat.info()
```

```python
bsmt_feat = bsmt_feat[bsmt_feat.isnull().any(axis=1)]
bsmt_feat
```

```python
bsmt_feat_all_nan = bsmt_feat[(bsmt_feat.isnull() | bsmt_feat.isin([0])).all(1)]# what is the meaning of this line
bsmt_feat_all_nan
```

```python
bsmt_feat_all_nan.shape
```

```python
qual = list (df.loc[:,df.dtypes == 'object'].columns.values)
qual
```

```python
# Filling the missing  value in bsmt features
for i in bsmt_col:
    if i in qual:
        bsmt_feat_all_nan[i] = bsmt_feat_all_nan[i].replace(np.nan,'NA')# replace the nan value by
    else:
        bsmt_feat_all_nan[i] = bsmt_feat_all_nan[i].replace(np.nan,0) # replace the nan value in
    
    bsmt_feat.update(bsmt_feat_all_nan) # update bsmt_feat df by bsmt_feat_all_nan
    
    df.update(bsmt_feat_all_nan)#update df by bsmt_feat_all_nan
    '''
    >>> df = pd.DataFrame({'A':[1,2,3],
    ...                       'B':[400,500,600]})
    >>>new_df = pd.DataFrame({'B':[4,5,6],
    ...                         'C':['7','8','9']})
    '''
    
```
