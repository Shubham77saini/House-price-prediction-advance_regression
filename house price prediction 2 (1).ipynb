```python
def category_onehot_multicols(multicolumns):
    final_df=df
    i=0
    for fields in multicolumns:
        print (fields)
        df1=pd.get_dummies(df[fields],drop_first=True)
        
        df.drop([fields],axis=1,inplace=True)
        if i == 0:
            final_df=df1.copy()
        else:
            final_df=pd.concat([final_df,df1],axis=1)
        i=i+1
        
    final_df=pd.concat([df,final_df],axis=1)
    return final_df
```

```python
df=category_onehot_multicols(columns)
```

```python
X_train=df_Train.drop(['SalePrice'],axis=1)
y_train=df_Train['SalePrice']
```

```python
from xgboost.sklearn import XGBRegressor
import xgboost
classifier=xgboost.XGBRegressor()
classifier.fit(X_train,y_train)
```

```python
import pickle
filename = 'final_model.pkl'
pickle.dump(classifier,open(filename,"wb"))
```
